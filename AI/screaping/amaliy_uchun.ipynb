```python
div= "<div class='city'>Toshkent</div>"
div2= "<div class='temp'>+15°C</div>"

soup=BeautifulSoup(div, "html.parser")
soup2=BeautifulSoup(div2, "html.parser")

print(soup.div.text)
print(soup2.div.text)
```

```python
html = """
<div class="city">Toshkent</div>
<div class="temp">+15°C</div>
<div class="city">Samarqand</div>
"""

soup=BeautifulSoup(html, "html.parser")

# birinchi sityni topish
first_city=soup.find("div", class_="city")
print(first_city.text)
```

```python
html = """
<div class="city">Toshkent</div>
<div class="temp">+15°C</div>
<div class="city">Samarqand</div>
<div class="city">Qashqadaryo</div>
"""

soup=BeautifulSoup(html, "html.parser")
# barcha citylarni topish
all_cities = soup.find_all("div", class_="city")
for city in all_cities:
    print(city.text)
```

```python
html ="""
<div class="city">Toshkent</div>
<div class="temp">+15°C</div>
<div class="city">Samarqand</div>
<div class="temp">+18°C</div>
"""
soup=BeautifulSoup(html, "html.parser")

first_c=soup.find("div", class_="city")
first_t=soup.find("div", class_="temp")

all_c=soup.find_all("div", class_="city")
all_t=soup.find_all("div", class_="temp")

print(first_c.text)
print(first_t.text)

for c in all_c:
    print(c.text)
for t in all_t:
    print(t.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.accuweather.com/uz/uz/qarshi/350541/weather-forecast/350541"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/143.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

print(response.status_code)  # 200 bo'lishi kerak

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    print(soup.prettify()[:1000])  # HTMLni tekshirish
else:
    print("Saytni ochib bo‘lmadi, status code:", response.status_code)
```
