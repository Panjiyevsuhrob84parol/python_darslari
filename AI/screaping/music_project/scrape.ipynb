```python
from email import header
from wsgiref import headers
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd

url = "https://www.last.fm/charts"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

start_time =time.time()
response =requests.get(url, headers=headers, timeout=10)
end_time=time.time()
total_time=end_time-start_time
print(f"time taken: {total_time}")
print(response.status_code)
soup = BeautifulSoup(response.text, "html.parser")

title=soup.title.text
print(title)
```

```python
print(soup.prettify)
```

```python
musics = soup.find_all("div", class_="charts-col")

print("bir nimalar")

for music in musics:
    text = music.get_text(separator="\n", strip=True)

    lines = text.split("\n")

    # Keraksiz bosh qatorlarni olib tashlaymiz
    clean_lines = [line for line in lines if line.strip() != ""]

    # print("aaaa: ", clean_lines[0])
    
    # Qolganlarini chiqaramiz
    for line in clean_lines[1:]:
        print("   ", line)

    print("-" * 50)
```

```python
top_tracks_col = soup.find("a", id="top-tracks").find_parent("div", class_="charts-col")

rows = top_tracks_col.select("tr.globalchart-item")

data = []

for row in rows:
    rank_el = row.select_one("td.globalchart-rank span")
    track_el = row.select_one("td.globalchart-name a")
    artist_el = row.select_one("td.globalchart-artist a")

    if not rank_el or not track_el or not artist_el:
        continue

    data.append({
        "rank": rank_el.get_text(strip=True),
        "track": track_el.get_text(strip=True),
        "artist": artist_el.get_text(strip=True)
    })

df = pd.DataFrame(data)
df
```
