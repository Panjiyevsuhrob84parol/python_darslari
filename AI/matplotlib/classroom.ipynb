```python
month=data["month_number"]
total_profit=data["total_profit"]
plt.plot(
    month,
    total_profit,
    linestyle="-.",
    color="blue",
    marker="*",
    markerfacecolor="green",
    linewidth="2",
    label="Total Profit"
)
# plt.xlabel="Month Number"
# plt.ylabel="Sold units number"
plt.legend(loc="upper left")
plt.show()
```

```python
months = data["month_number"]
import matplotlib.pyplot as plt
plt.plot(months, data["facecream"], label="Face Cream", marker='o', color='red')
plt.plot(months, data["facewash"], label="Face Wash", marker='D', color='green')
plt.plot(months, data["toothpaste"], label="Toothpaste", marker = 's', color='yellow')
plt.plot(months, data["bathingsoap"], label="Bathing Soap", marker = 's', color='blue')
plt.plot(months, data["shampoo"], label="Shampoo", marker = '*', color='purple')
plt.plot(months, data["moisturizer"], label="Moisturizer", marker = 'o', color='black')

plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(loc="upper left")

plt.show()
```
